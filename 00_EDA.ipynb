4.1.2. What percentage of orders are shipped on time (On-Time Shipping Percentage)?


```python
df["OTSP"] = df["shipschedule"] - df["OFCT"]

on_time_shipping = df[df["OTSP"] >= 0]
on_time_shipping_pct = on_time_shipping["OTSP"].count() / df["OTSP"].count()*100
print("The Overall On-Time Shipping Percentage From 2015 - 2017 is:",on_time_shipping_pct,"%")
```

```python
mod_full = df.groupby("modename")["OTSP"].count()
mod_real = on_time_shipping.groupby("modename")["OTSP"].count()
mod_joined = pd.merge(mod_full, mod_real,left_on='modename', right_on='modename', how='left')
mod_joined["on_time_shipping_pct"] = mod_joined["OTSP_y"]/ mod_joined["OTSP_x"]*100

mod_joined
```

```python
mod_full = df.groupby("whcountry")["OTSP"].count()
mod_real = on_time_shipping.groupby("whcountry")["OTSP"].count()
mod_joined = pd.merge(mod_full, mod_real,left_on='whcountry', right_on='whcountry', how='left')
mod_joined["on_time_shipping_pct"] = mod_joined["OTSP_y"]/ mod_joined["OTSP_x"]*100

mod_joined
```

```python
mod_full = df.groupby("custregion")["OTSP"].count()
mod_real = on_time_shipping.groupby("custregion")["OTSP"].count()
mod_joined = pd.merge(mod_full, mod_real,left_on='custregion', right_on='custregion', how='left')
mod_joined["on_time_shipping_pct"] = mod_joined["OTSP_y"]/ mod_joined["OTSP_x"]*100

mod_joined_pct = mod_joined["on_time_shipping_pct"].sort_values()
mod_joined_pct.plot(kind = "barh")
```

```python
mod_full = df.groupby("dept")["OTSP"].count()
mod_real = on_time_shipping.groupby("dept")["OTSP"].count()
mod_joined = pd.merge(mod_full, mod_real,left_on='dept', right_on='dept', how='left')
mod_joined["on_time_shipping_pct"] = mod_joined["OTSP_y"]/ mod_joined["OTSP_x"]*100

mod_joined_pct = mod_joined["on_time_shipping_pct"].sort_values()
mod_joined_pct.plot(kind = "barh")
```

```python
mod_full = df.groupby("orderqty")["OTSP"].count()
mod_real = on_time_shipping.groupby("orderqty")["OTSP"].count()
mod_joined = pd.merge(mod_full, mod_real,left_on='orderqty', right_on='orderqty', how='left')
mod_joined["on_time_shipping_pct"] = mod_joined["OTSP_y"]/ mod_joined["OTSP_x"]*100

mod_joined_pct = mod_joined["on_time_shipping_pct"].sort_values()
mod_joined_pct.plot(kind = "barh")
```

4.2. Sales

4.2.1. Which product dept, order quantity generate the most revenue and profit?

```python
#Assuming that "profit" term refers to "gross profit"
data["netsales"] = data["grosssales"]*(1-data["discpc"])
df = data[data["netsales"] >= data["profit"]]
df["profitmargin"] = df["profit"]/df["netsales"]
```

```python
#profit margin by product department from 2015 - 2017
df.groupby("dept")["profitmargin"].mean().sort_values().plot(kind = "barh")
```

```python
#profit margin by order quantity from 2015 - 2017
df.groupby("orderqty")["profitmargin"].mean().sort_values().plot(kind = "barh")
```

4.2.2. What is the overall sales growth rate and profit margin trend over time? Are there seasonal patterns?

```python
#profit margin trend during 2015 - 2017 period
df["ordermonth"] = df["ordertime"].dt.month
df["orderyear"] = df["ordertime"].dt.year
```

```python
df_group_time = df.groupby(["ordermonth"])["profitmargin"].mean().plot(y="profitmargin", kind="line") 
plt.show() 
```

```python
df_group_time = df.groupby(["orderyear","ordermonth"])["profitmargin"].mean().plot(y="profitmargin", kind="line") 
plt.show() 
```

```python
#net sales growth rate from 2015 - 2017
df["ordermonth"] = df["ordertime"].dt.month
df["orderyear"] = df["ordertime"].dt.year
df_group_time = df.groupby(["orderyear","ordermonth"])[["netsales"]].sum()

df_sales = df_group_time.reset_index()
df_sales["salesgrowthrate"] = df_sales["netsales"].pct_change()
```

```python
df_sales.groupby(["ordermonth"])["salesgrowthrate"].mean().plot(x = "orderyear",y="salesgrowthrate", kind="line") 
plt.show()
```

```python
df_sales.groupby(["orderyear","ordermonth"])["netsales"].mean().plot(x = "orderyear",y="netsales", kind="line") 
plt.show()
```

4.2.3.How do discounts impact sales volume and profitability?


```python
#correlation of sales and dispc
df_group_time = df.groupby("discpc")["netsales"].sum()
df_sales = df_group_time.reset_index()
df_sales["salesgrowthrate"] = df_sales["netsales"].pct_change()
```

```python
sns.jointplot(data=df_sales, x='discpc', y='salesgrowthrate', kind='reg')
```

```python
#correlation of discpc and profit margin
df_group_time = df.groupby("discpc")[["profit","netsales"]].sum()
df_profit = df_group_time.reset_index()
df_profit["profitmargin"] = df_profit["profit"]/df_profit["netsales"]
```

```python
sns.jointplot(data=df_profit, x='discpc', y='profitmargin', kind='reg')
```

4.2.4. What is the relationship between supply chain costs and sales revenue (Inventory-to-Sales Ratio)?

```python
inv = pd.read_csv("inventory.csv")
inv
```

```python
#inventory-to-sales ratio by month
inv["invcost"] = inv["whinv"]*inv["invcostperunit"]
df_inv = inv.groupby(["year","month"])["invcost"].sum().reset_index()

df_netsales = df.groupby(["orderyear","ordermonth"])["netsales"].sum().reset_index()

inv_to_sales = pd.merge(df_inv, df_netsales,left_on=["year","month"], right_on=["orderyear","ordermonth"], how='left')
inv_to_sales["invtosales"] = inv_to_sales["invcost"]/inv_to_sales["netsales"]
```

```python
inv_to_sales.groupby(["month"])["invtosales"].mean().plot(y="invtosales", kind="line") 
plt.show()
```
